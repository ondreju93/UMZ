$$ RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^n\left(y_i-\hat{y}_i\,\right)^2} $$

```python
import csv
reader = csv.reader(open("train.csv"), delimiter=",")
next(reader, None)

#Id,Expected,Rooms,SqrMeters,Floor,Location,Description
ID = list()
expected = list()
rooms = list()
sqr = list()
floor = list()
loc = list()
desc = list()
minus_sqrt_floor_log = list()
minusfloor = list()

for ii, ei, ri, si, fi, li, di in reader:
    ID.append(int(ii))
    expected.append(int(ei))
    rooms.append(int(ri))
    sqr.append(float(si))
    floor.append(int(fi))
    loc.append(li)
    desc.append(di)
    minus_sqrt_floor_log.append(-sqrt(np.log(float(fi))))
    minusfloor.append(-float(fi))

def regdots(x, y, lbl='x~y', xlbl='x', ylbl='y', color='g', fig=pl.figure(figsize=(16*.6,9*.6)) ,subplot=111,
            xlim=[0, 100], ylim=[50000, 1500000], yTickVisible=False):    
    ax = fig.add_subplot(subplot)
    fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)
    ax.scatter(x, y, c=color, s=80, label=lbl)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_xlabel(xlbl)
    ax.set_ylabel(ylbl)
    ax.margins(.05,.05)
    pl.setp(ax.get_yticklabels(), visible=yTickVisible)
    return fig

def legend(fig):
    ax = fig.axes[0]
    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, fontsize='15', loc='lower right')

# Funkcja fun jest argumentem
def regline(fig, fun, theta, x):
    ax = fig.axes[0]
    x0, x1 = min(x), max(x)
    X = [x0, x1]
    Y = [fun(theta, x) for x in X]
    ax.plot(X, Y, linewidth='2', 
            label=(r'$y=%.2f+%.2f x$' % (theta[0], theta[1])))
    
# Definicja prostej, funkcja jednej zmienniej

def h(theta, x):
    return theta[0] + theta[1]*x

# definicja funkcji błędu

def J(h, theta, x, y):
    m = len(y)
    return (1.0/(2*m) * sum((h(theta, x[i]) - y[i])**2 
                          for i in range(m)))
fig1 = regdots(floor,expected, lbl='price ~ floor', xlbl='floor', ylbl='expected price', subplot=131,
               xlim=[0, 18], ylim=[50000, 1500000], yTickVisible=True)
fig2 = regdots(rooms,expected, lbl='price ~ rooms', xlbl='rooms', ylbl='expected price', fig=fig1,subplot=132,
               xlim=[0, 12], ylim=[50000, 1500000], color='red')
fig3 = regdots(sqr,expected, lbl='price ~ square meters', xlbl='sqr meters', ylbl='expected price', fig=fig2,subplot=133,
               xlim=[0, 200], ylim=[50000, 1500000], color='yellow')
pl.show()
```

A. Za pomocą macierzy normalnej lub innej metody optymalizacyjnej zbuduj model regresyjny dla pierwszej większej cześci.  
B. Zastosuj tak stworzony model (parametry $\theta$) do pozostałych 1000 wierszy danych.  
C. Oblicz RMSE dla drugiej cześci danych.  
D. Eksperymentuj z doborem cech, aby uzyskać jak najmniejszy RMSE (Oprócz oczywistych cech numerycznych jak metraż, cechy mogą być kombinacjami innych cech, lub symbolicznymi cechami opisującymi występowanie wyrazów, itp.)

```python
#Id,Expected,Rooms,SqrMeters,Floor,Location,Description
      
# dane treningowe
TID = np.array(ID[:4000])
Texpected = np.array(expected[:4000])
Trooms = np.array(rooms[:4000])
Tsqr = np.array(sqr[:4000])
Tminus_sqrt_floor_log = np.array(minus_sqrt_floor_log[:4000])
Tfloor = np.array(floor[:4000])
Tfloorlog = np.log(Tfloor)
Tloc = np.array(loc[:4000])
Tdesc = np.array(desc[:4000])

# dane do testow
ID_test = np.array(ID[4000:5000])
expected_test = np.array(expected[4000:5000])
rooms_test = np.array(rooms[4000:5000])
sqr_test = np.array(sqr[4000:5000])
minus_sqrt_floor_log_test = np.array(minus_sqrt_floor_log[4000:5000])
floor_test = np.array(floor[4000:5000])
floorlog_test = np.log(floor_test)
loc_test = np.array(loc[4000:5000])
desc_test = np.array(desc[4000:5000])
      
TMx = np.matrix(np.concatenate((np.array([one[0] for one in np.ones((4000, 1))]).reshape(4000, 1),
                                Trooms.reshape(4000, 1),
                                Tsqr.reshape(4000, 1),
                                Tminus_sqrt_floor_log.reshape(4000,1)), axis=1).reshape(4000, 4))
      
yMx = np.matrix(Texpected.reshape(4000, 1))
      
def norm(X,y):
    return (X.T*X)**-1*X.T*y
      
def hMx(theta, X):
    return X*theta

def RMSE(values, predicted):
    temptable = (np.asarray(values).flatten() - np.asarray(predicted).flatten())**2
    return np.sqrt(np.sum(temptable)/values.size)

thetaModel = norm(TMx, yMx)
      
print 'obliczone za pomoc\u0105 macierzy normalnej, parametry theta:'
print thetaModel
      
      
      
Mx_Test = np.matrix(np.concatenate((np.array([one[0] for one in np.ones((1000, 1))]).reshape(1000, 1),
                                    rooms_test.reshape(1000, 1),
                                    sqr_test.reshape(1000, 1),
                                    minus_sqrt_floor_log_test.reshape(1000,1)), axis=1).reshape(1000, 4))
      
predicted = hMx(thetaModel, Mx_Test)
Results = np.concatenate((ID_test.reshape(1000, 1), expected_test.reshape(1000, 1), predicted.reshape(1000, 1)), axis=1)
      
writer = csv.writer(open("result.csv", 'w'), delimiter =',')
writer.writerow(('ID', 'Expected', 'Predicted'))
writer.writerows(Results.tolist())

error = RMSE(expected_test, predicted)      
print 'obliczony blad dla drugiej czesci danych:'
print error

#942230 - najmniejszy blad byl przy dwoch cechach  - liczbie pokoi i metrazu
```
