```python
# Nagłówki

import numpy as np
import matplotlib
import matplotlib.pyplot as pl
import IPython.html.widgets as widgets
from IPython.html.widgets import fixed
from IPython.display import display

import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

from IPython.display import display, Math, Latex

def LatexMatrix(matrix):
    ltx = r'\left[\begin{array}'
    m, n = matrix.shape
    ltx += '{' + ("r" * n) + '}'
    for i in range(m):
        ltx += r" & ".join([('%.4f' % j.item()) for j in matrix[i]]) + r" \\ "
    ltx += r'\end{array}\right]'
    return ltx
```

3. Dodaj dowolną prostą z ręcznie wybranymi parametrami θ, oblicz wartość jej błędu J(θ).

```python
import csv
reader = csv.reader(open("fires_thefts.txt"), delimiter=",")

# x- fires
x = list()
# y - thefts
y = list()

for xi, yi in reader:
    x.append(float(xi))
    y.append(float(yi))

def regdots(x, y):    
    fig = pl.figure(figsize=(16*.6,9*.6))
    ax = fig.add_subplot(111)
    fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)
    ax.scatter(x, y, c='g', s=80, label="Dane")
    
    ax.set_xlabel("Nr of fires")
    ax.set_ylabel("Nr of thefts")
    ax.margins(.05,.05)
    pl.ylim(min(y)-1, max(y)+1)
    pl.xlim(min(x)-1, max(x)+1)
    return fig

def legend(fig):
    ax = fig.axes[0]
    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, fontsize='15', loc='lower right')

# Funkcja fun jest argumentem
def regline(fig, fun, theta, x):
    ax = fig.axes[0]
    x0, x1 = min(x), max(x)
    X = [x0, x1]
    Y = [fun(theta, x) for x in X]
    ax.plot(X, Y, linewidth='2', 
            label=(r'$y=%.2f+%.2f x$' % (theta[0], theta[1])))
    
# Definicja prostej, funkcja jednej zmienniej

def h(theta, x):
    return theta[0] + theta[1]*x

# definicja funkcji błędu

def J(h, theta, x, y):
    m = len(y)
    return (1.0/(2*m) * sum((h(theta, x[i]) - y[i])**2 
                          for i in range(m)))

fig = regdots(x,y)
theta0 = 30
theta1 = 1.5
regline(fig, h, [theta0, theta1], x)
legend(fig)

display(Math(r'\large\quad \theta = ' + 
             LatexMatrix(np.matrix([theta0, theta1]).reshape(2,1))))

display(Math(r"\Large J(\theta) = %.4f" % J(h,[theta0, theta1],x,y)))
```

4.Oblicz parametry θ krzywej regresyjnej za pomocą macierzy normalnej, przedstaw na nowym wykresie.

```python
# Wczytywanie danych do postaci macierzy

data = np.loadtxt("fires_thefts.txt", delimiter=",")
m, np1 = data.shape # np1 : n plus 1
n = np1 - 1
Xn = data[:,0:n].reshape(m,n)

# Dodaj kolumne jedynek do macierzy
XMx = np.matrix(np.concatenate((np.ones((m, 1)), Xn), axis=1)).reshape(m,np1)

yMx = np.matrix(data[:,1]).reshape(m,1)

def hMx(theta, X):
    return X*theta

# Rysowanie wykresów dla danych macierzowych

def regdotsMx(X, y):    
    fig = pl.figure(figsize=(16*.6,9*.6))
    ax = fig.add_subplot(111)
    fig.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)
    ax.scatter(X[:,1], y, c='g', s=80, label="Dane")
    
    ax.set_xlabel("nr of fires")
    ax.set_ylabel("nr of thefts")
    ax.margins(.05,.05)
    pl.ylim(y.min()-1, y.max()+1)
    pl.xlim(np.min(X[:,1])-1, np.max(X[:,1])+1)
    return fig

def reglineMx(fig, fun, theta, X):
    ax = fig.axes[0]
    x0, x1 = np.min(X[:,1]), np.max(X[:,1])
    L = [x0, x1]
    LX = np.matrix([1, x0, 1, x1]).reshape(2,2)
    ax.plot(L, fun(theta, LX), linewidth='2', 
            label=(r'$y=%.4f+%.4f x$' % (theta[0][0], theta[1][0])));
    
fig = regdotsMx(XMx, yMx)

# Wersja macierzowa funkcji kosztu

def JMx(theta,X,y):
    m = len(y)
    J = 1.0/(2.0*m)*((X*theta-y).T*(X*theta-y))
    return J.item()

def norm(X,y):
    return (X.T*X)**-1*X.T*y

thetaNorm = norm(XMx, yMx)
display(Math(r'\Large \theta = ' + LatexMatrix(thetaNorm)))
display(Math(r'\Large J(\theta) = %.4f' % JMx(thetaNorm, XMx, yMx)))

reglineMx(fig, hMx, thetaNorm, XMx)
legend(fig)
```

5.Oblicz parametry θ krzywej regresyjnej za pomocą algorytmu gradientów prostych (Gradient Descent, GD), można wybrać wersję iteracyjną lub macierzową) i przedstaw na nowym wykresie.

```python
# Wersja macierzowa gradientu funkcji kosztu

def dJMx(theta,X,y):
    return 1.0/len(y)*(X.T*(X*theta-y)) 

# Implementacja algorytmu gradient descent za pomocą numpy i macierzy

def GDMx(fJ, fdJ, theta, X, y, alpha=0.1, eps=10**-3):
    errorCurr = fJ(theta, X, y)
    errors = [[errorCurr, theta]]
    while True:
        theta = theta - alpha * fdJ(theta, X, y) # implementacja wzoru
        errorCurr, errorPrev = fJ(theta, X, y), errorCurr
        if abs(errorPrev - errorCurr) <= eps:
            break
        errors.append([errorCurr, theta]) 
    return theta, errors

thetaStartMx = np.matrix([0,0]).reshape(2,1)

# Zmieniamy wartosci alpha (rozmiar kroku) oraz eps (kryterium stopu)
thetaBestMx, errors = GDMx(JMx, dJMx, thetaStartMx, XMx, yMx, alpha=0.001, eps=0.0000001)

################################################################
display(Math(r'\large\textrm{Wynik:}\quad \theta = ' + 
             LatexMatrix(thetaBestMx)))  

fig = regdotsMx(XMx, yMx)
reglineMx(fig, hMx, thetaBestMx, XMx)
legend(fig)
```

6.Dla różnych wartości α∈{0.001,0.01,0.1} (α to długość kroku) stwórz wykres, który ilustruje progresję wartości J(θ) dla pierwszych 200 króków algorytmu GD. (Wykres ma zatem na osi x kroki od 0 do 200, na osi y wartosci J(θ), oraz przedstawia trzy krzywe, po jednej dla każdej wartości α.)

```python
def GDMx(fJ, fdJ, theta, X, y, alpha=0.1, eps=10**-3, maxSteps=1000):
    errorCurr = fJ(theta, X, y)
    errorsPrim = [[errorCurr, theta]]
    i = 0
    while True:
        theta = theta - alpha * fdJ(theta, X, y) # implementacja wzoru
        errorCurr, errorPrev = fJ(theta, X, y), errorCurr
        if abs(errorPrev - errorCurr) <= eps:
            break
        
        errorsPrim.append([errorCurr, theta]) 
        i+=1
        if i >= maxSteps:
            break
    return theta, errorsPrim

    
## initialize the axes
fig = pl.figure(figsize=(12, 10))
ax = fig.add_subplot(111)

## format axes
ax.set_ylabel('J')
ax.set_xlabel('step')
ax.set_title('J ~ step')


_, errors1 = GDMx(JMx, dJMx, thetaStartMx, XMx, yMx, alpha=0.1, eps=0.0001)
line1, = ax.plot(range(200), [e[0] for e in errors1[:200]], color='red', lw=10, label='alfa 0.01')
_, errors01 = GDMx(JMx, dJMx, np.matrix([0,0]).reshape(2,1), XMx, yMx, alpha=0.01, eps=0.0001)
line01, = ax.plot(range(200), [e[0] for e in errors01[:200]], color='blue', lw=2, label='alfa 0.001')
_, errors001 = GDMx(JMx, dJMx, np.matrix([0,0]).reshape(2,1), XMx, yMx, alpha=0.001, eps=0.0001)
line001, = ax.plot(range(200), [e[0] for e in errors001[:200]], color='orange', lw=1, label='alfa 0.0001')


legend(fig)
pl.axis([0, 200, 0, 10000])
pl.show()
```
